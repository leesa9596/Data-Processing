(1) 사고유형대분류가 '차대차'인 사고유형별 사망비율을 막대그래프로 시각화하시오.(사망비율은 사망자수/사고건수)

(2) 사고유형대분류별로 사망자수가 가장 많은 사고 유형을 출력

(3) 사고유형별 중상비율을 계산하고 중상비율인 50%가 넘는 데이터를 중상비율.csv 로 저장하시오.

```python
#그래프 폰트 설정
from matplotlib import pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'
```

```python
import pandas as pd
df = pd.read_excel('accident_stat.xlsx')
```

```python
df
```

```python
df_cartocar = df[df['사고유형대분류']=='차대차'].copy()
df_cartocar['사망비율'] = df_cartocar['사망자수']/df_cartocar['사고건수']
```

```python
df_cartocar
```

```python
plt.bar(df_cartocar['사고유형'], df_cartocar['사망비율'])
plt.show()
```

```python
df.loc[df.groupby('사고유형대분류')['사망자수'].idxmax(),'사고유형'].tolist()
```

```python
df['중상비율'] = df['중상자수']/df['사고건수']
df
```

```python
df_50 = df[df['중상비율']>0.5]
df_50
```
